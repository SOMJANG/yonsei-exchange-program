```python
def access_univ_info(dataframe, univ_code):
    df_row = dataframe.loc[dataframe["대학코드"] == univ_code]
    str_univ_info = df_row["gen_info"].to_string(index=False).lstrip()
    return str_univ_info 
```

```python
# removing junks from the string data
def remove_junk(str_data):
    str_data = str_data.replace("\\\\n", "")
    str_data = str_data.replace("\\n", "")
    str_data = str_data.replace("\n", "")
    str_data = str_data.replace("\\", "")
    str_data = str_data.replace("\\t", "")

    # replace multiple spaces into single space
    str_data = ' '.join(str_data.split())
    return str_data
```

```python
# collecting all column values into one string
def column_to_string(dataframe, column_name: str):
    str_corpus = dataframe[column_name].to_string(index=False).lstrip()
    str_corpus = remove_junk(str_corpus)
    return str_corpus
```

## Tokenization using konlpy for column values
## konlpy 모듈로 형태소분석, 형태소 등장 빈도수 확인

```python
# konlpy 기반 tokenizer 갖고 오기
from konlpy.tag import Okt
Okt_tokenizer = Okt()
example_pos_okt = Okt_tokenizer.pos(str_example, norm=True, stem=True)
example_pos_okt[:10]
```

```python
# konlpy 기반 tokenizer 갖고 오기
from konlpy.tag import Mecab
mecab_tokenizer = Mecab()
example_pos_mecab = mecab_tokenizer.pos(str_example, flatten=True)
example_pos_mecab[:10]
```

```python
# collecting all column values into one string
gen_corpus = column_to_string(df_specific, "gen_info")
print("글자수:", len(gen_corpus))
gen_corpus[:1000]
```

```python
# tokenizing sentences into nouns, verbs, etc
# tokenization for mecab: http://openuiz.blogspot.com/2016/07/mecab-ko-dic.html
%time mecab_gen_info = mecab_tokenizer.pos(gen_corpus, flatten=True)
print(len(mecab_gen_info))
mecab_gen_info[:10]
```

```python
# collect only nouns into list
%time mecab_gen_nouns = mecab_tokenizer.nouns(''.join(gen_corpus))
print(len(mecab_gen_nouns))
mecab_gen_nouns[:10]
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# mecab으로 추출한 명사로 워드클라우드를 그린다.
backgroundcolor = "white"
width=800
height=600
wordcloud = WordCloud(font_path = fontpath,
                        # stopwords = STOPWORDS, 
                        background_color = backgroundcolor, 
                        max_words= 50,
                        max_font_size= 300,
                        width = width, 
                        height = height)

# 빈도수 별로 단어를 크게 표시한다
wordcloud.generate_from_frequencies(frequencies = mecab_gen_nouns_counted)
plt.figure(figsize = (15 , 10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show() 
```
