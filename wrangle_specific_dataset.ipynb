```python
# summarizing above process into function
df_univ = pd.read_csv(f"./data_wrangled/df_univ_wrangled.csv", encoding="utf-8")

def make_univ_corpus(df_univ, univ_code):
    
    # getting series based on university code
    series_row = df_univ.loc[df_univ["대학코드"] == univ_code]

    # getting univ name and country in series
    univ_name = series_row["대학명"].to_string(index=False).lstrip()
    country = series_row["국가"].to_string(index=False).lstrip()

    df_temp = pd.read_csv(f"{SOURCE_SPECIFIC}/{univ_code}_reviews_specific.csv", encoding="utf-8")

    # makign dictionary with same keys to make dataframe
    column_dict = {}
    column_dict["대학코드"] = univ_code
    column_dict["대학명"] = univ_name

    # combining specific reviews into one corpus according to keys
    list_specific_columns = df_temp.columns.to_list()
    for column_item in list_specific_columns:
        column_corpus = df_temp[column_item].to_string(index=False)
        
        # replace unnecassary stuffs
        column_corpus = column_corpus.replace("\\", "")
        column_corpus = column_corpus.replace("\n", "")
        column_corpus = column_corpus.replace("\\n", "")
        column_corpus = column_corpus.replace("\\t", "")
        
        # replace multiple spaces into single space
        column_corpus = ' '.join(column_corpus.split())

        column_dict[column_item] = column_corpus
    return column_dict
```

```python
df_ever_been = df_univ.dropna(subset=["NUM_VISITS"])
df_ever_been.head()
```

```python
list_univ_code = df_ever_been["대학코드"].to_list()
list_univ_code[:10]
```

```python
list_dict_corpus = []
for university in list_univ_code:
    dict_temp = make_univ_corpus(df_ever_been, university)
    list_dict_corpus.append(dict_temp)
```
